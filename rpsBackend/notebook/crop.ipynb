This Notebook is meant to use HTD to crop any dataset images with just the hand if the person is in it. It uses the hand tracking model to crop the image and save it in the same folder as the original image.

Ensure the parent folder has folders with rock, paper, scissors

```python
import cv2
import mediapipe as mp
import math as math
import HandTrackingDynamic as htd
import numpy as np
import os
```

```python
detector = htd.HandTrackingDynamic()

for dirname, _, filenames in os.walk(r'C:\Users\micha\Downloads\archive\test'):
    for filename in filenames:
        filepath = os.path.join(dirname, filename)
        #read image from directory using Image
        img = cv2.imread(filepath)
        frame = detector.findFingers(img, draw=False)
        lmsList, bbox = detector.findPosition(frame, handDraw=False)
        if len(bbox) == 0:
            os.remove(filepath)
            print("Deleted: ", filepath)
            continue
        xmin, ymin, xmax, ymax = bbox

        img = img[max(0, ymin - 25):ymax + 25, max(0, xmin - 25):xmax + 25]
        
        cv2.imwrite(filepath, img) 
        print("Cropped: ", filepath)
        
```
